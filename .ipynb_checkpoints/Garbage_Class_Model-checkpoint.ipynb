```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
import zipfile
import os

# Path to the uploaded .zip file
zip_file_path = 'garbage_classification.zip'

# Directory to unzip the files to
extract_dir = 'garbage_classification'

# Unzipping the folder
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# Optional: Remove the .zip file after extraction
os.remove(zip_file_path)
```

```python
data_train_path = 'garbage_classification/garbage_classification/train'
data_test_path = 'garbage_classification/garbage_classification/test'
data_val_path = 'garbage_classification/garbage_classification/validation'
```

```python
img_width = 180
img_height = 180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle = True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False)
```

```python
data_cat = data_train.class_names
```

```python
print(data_cat)
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    shuffle = True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False)
```

```python
data_test = tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    shuffle = True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False)
```

```python
plt.figure(figsize=(10,10))
for image, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
```

```python
data_train
```

```python
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_cat))
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
epochs_size = 25
history = model.fit(data_train, validation_data = data_val, epochs=epochs_size)
```

```python
 epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range, history.history['accuracy'], label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'], label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range, history.history['loss'], label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'], label = 'Validation Loss')
plt.title('Loss')
```

```python
image = 'brown-glass505.jpg'
image = tf.keras.utils.load_img(image, target_size=(img_height, img_width))
img_arr = tf.keras.utils.array_to_img(image)
img_bat=tf.expand_dims(img_arr,0)
```

```python
predict = model.predict(img_bat)
```

```python
score = tf.nn.softmax(predict)
```

```python
print('Veg/Fruit in image is {} with accuracy of {:0.2f}'.format(data_cat[numpy.argmax(score)],numpy.max(score)*100))
```
